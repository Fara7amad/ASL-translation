```python
import cv2
import numpy as np
import mediapipe as mp
import tensorflow as tf
import os
```

```python
model = tf.keras.models.load_model('final_model.keras')
```

```python
classes = sorted(os.listdir('asl_alphabet_train/asl_alphabet_train'))
```

```python
# Initialize MediaPipe Hands
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Function to extract landmarks from a single frame
def extract_landmarks_from_frame(frame):
    landmarks = []
    with mp_hands.Hands(static_image_mode=False, max_num_hands=1, min_detection_confidence=0.7, min_tracking_confidence=0.5) as hands:
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        result = hands.process(frame_rgb)
        if result.multi_hand_landmarks:
            for hand_landmarks in result.multi_hand_landmarks:
                mp_drawing.draw_landmarks(frame, hand_landmarks, mp_hands.HAND_CONNECTIONS)
                for lm in hand_landmarks.landmark:
                    landmarks.append([lm.x, lm.y, lm.z])
        if landmarks:
            return np.array(landmarks).flatten()
        else:
            return np.zeros(63)  # 21 landmarks with x, y, z coordinates

# Initialize the webcam
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        landmarks = extract_landmarks_from_frame(frame)

        # Make a prediction if landmarks are detected
        if landmarks.any():
            landmarks = np.reshape(landmarks, (1, -1))  # Reshape for model input
            prediction = model.predict(landmarks)
            predicted_class = np.argmax(prediction)
            predicted_label = classes[predicted_class]

            # Display the predicted label on the frame
            cv2.putText(frame, predicted_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

        # Display the resulting frame
        cv2.imshow('Sign Language Translation', frame)

        # Break the loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the capture and close the windows
cap.release()
cv2.destroyAllWindows()
```
